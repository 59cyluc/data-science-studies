```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canadian_immegration_data.csv")
df.head()
```

```python
df.info()
```

```python
df.query("Country == 'Brazil'")
```

```python
df.set_index("Country", inplace=True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brazil = df.loc["Brazil", anos]
brazil.head()
```

```python
brazil_dict = {"ano": brazil.index.tolist(), "imigrantes": brazil.values.tolist()}
dados_brazil = pd.DataFrame(brazil_dict)
dados_brazil.head()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brazil["ano"], dados_brazil["imigrantes"], lw=3, color="g")
ax.set_title("Imigracao do Brazil para o Canada\n1980 a 2013", fontsize=18, loc="left")
ax.set_xlabel("Ano", fontsize=14)
ax.set_ylabel("Numero de imigrantes", fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle="--")
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].plot(dados_brazil["ano"], dados_brazil["imigrantes"], lw=3, color="g")
axs[0].set_title("Imigracao do Brazil para o Canada\n1980 a 2013")
axs[0].set_xlabel("Ano")
axs[0].set_ylabel("Numero de imigrantes")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brazil["imigrantes"])
axs[1].set_title("Boxplot da imigracao do Brazil para o Canada\n1980 a 2013")
axs[1].set_xlabel("Brazil")
axs[1].set_ylabel("Numero de imigrantes")
axs[1].grid()
```

```python
dados_brazil_argentina = df.loc[["Brazil", "Argentina"], anos]
print(dados_brazil_argentina)
dados_brazil_argentina = dados_brazil_argentina.T
dados_brazil_argentina.head()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brazil_argentina["Brazil"], label="Brazil")
ax.plot(dados_brazil_argentina["Argentina"], label="Argentina")
ax.set_title("Imigracao do Brazil e Argentina para o Canada\n1980 a 2013")
ax.set_xlabel("Ano")
ax.set_ylabel("Numero de imigrantes")
ax.xaxis
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle("Imigração dos quatro maiores países da América do Sul para o Canadá\n1980 a 2013")

axs[0, 0].plot(df.loc["Brazil", anos])
axs[0, 0].set_title("Brazil")

axs[0, 1].plot(df.loc["Colombia", anos])
axs[0, 1].set_title("Colombia")

axs[1, 0].plot(df.loc["Argentina", anos])
axs[1, 0].set_title("Argentina")

axs[1, 1].plot(df.loc["Peru", anos])
axs[1, 1].set_title("Peru")

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel("Ano")
    ax.set_ylabel("Número de imigrantes")
    ax.grid(linestyle="-")

ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
america_sul = df.query("Region == 'South America'")
america_sul.head()
```

```python
america_sul_sorted = america_sul.sort_values("Total", ascending=True)
```

```python
cores = list()
for pais in america_sul_sorted.index:
    if pais == "Brazil":
        cores.append("green")
    else:
        cores.append("silver")
fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted["Total"], color=cores)
ax.set_title("America do Sul: Brasil foi o quarto pais com mais imigrantes para o Canada\n1980 a 2013", fontsize=18, loc="left")
ax.set_xlabel("Numero de imigrantes", fontsize=14)
ax.set_ylabel("")
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted["Total"]):
    ax.text(v + 20, i, str(v), color="black", fontsize=10, ha="left", va="center")

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis="both", which="both", length=0)
fig.savefig("imigracao_america_sul_canada.png", transparent=False, dpi=300, bbox_inches="tight")
plt.show()
```
